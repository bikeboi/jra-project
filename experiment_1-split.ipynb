# Experiment 1 - Two Class Supervised & Unsupervised

## Details
Observe the effect of selective PN to iKC connectivity on the network's representational power.

### Model Info
**Random Sparse**: Neurons are connected randomly with probabilty $p = 0.05$ across the network. Network parameters:
\begin{align}
& g_{PN,iKC} \in \mathcal{N}(0.5, 0.1)\\
& g_{iKC,eKC} \in \mathcal{N}(1.25, 0.5)\\
& \tau_{PN,iKC} = \tau_{iKC,eKC} = 5\\
& p_{PN,iKC} = p_{iKC,eKC} = 0.05
\end{align}

All neurons are current-based LIFs with exponential post-synaptic decay. The following parameters are of note:
\begin{align}
\tau_m &= 10\:\mathrm{ms}\\
v_{thresh} &= -50\:\mathrm{mV}
\end{align}

The iKC and eKC populations form sWTA circuits by lateral inhibition:

Neuron population sizes are as follows:

\begin{align}
& N_{PN} = \frac{d_{input}}{5}\\
& N_{iKC} = 10 \times N_{PN}\\
& N_{eKC} = 500
\end{align}

where $d_{input}$ is the dimensionality of (flattened) image tensors.

Each input class contains 20 samples, and each sample is shown $N_{round}$ times to the model, for a total sample size of $N_{sample} = N_{round} \times N_{class}$. The number of samples per class is kept balanced.

### Spike Coding
Prior to spike coding, images are downscaled by a factor of 4. The images are binarized by setting the top 5% of active pixels to 1, and the rest to 0. This also goes to create sparser input patterns. Images are encoded into spike times as instantaneous snapshots plus a small amount of jitter drawn from $\mathcal{U}(-\epsilon, \epsilon)$. During training, samples drawn are shown to the model within 50ms windows. eKC neuron activity, as a response to an input pattern, is the firing rate within these windows.

To analyse eKC population activity, we calculate:
- Population distance tables (cosine and manhattan distance)
- Changes in mean pre-synaptic eKC weights over time
- Per interval per eKC neuron firing rate and input class correlation
- Per interval mean (eKC) firing rate and input class correlation

## eKC activity

```python
eKC_activity = []
for i in range(RUNS):
    activity = calculate_activity(
        data['eKC'].segments[i].spiketrains, 
        sim_params['intervals'], 
        sim_params['t_snapshot']
    )
    
    eKC_activity.append(activity)

eKC_activity = np.array(eKC_activity) / sim_params['t_snapshot']
```

```python
def distance_matrix(activity, metric):
    return np.nan_to_num(scipy.spatial.distance.squareform(scipy.spatial.distance.pdist(activity, metric=metric)))

# Average across runs
run_mean_activity = eKC_activity.mean(axis=0)

# Plot distance tables
sns.set_style("white")
plt.gcf().set_size_inches(10,10)

plt.imshow(distance_matrix(run_mean_activity, 'cosine'), cmap='viridis')
plt.title("Cosine Distance Matrix")
plt.colorbar()
plt.show()
```

```python
run_mean_activity.shape
```

```python
labels = sim_params['labels'][1:]
population_mean = run_mean_activity.mean(axis=0)

# Spearman correlation
pop_corr_trace = []
for i in np.arange(2, labels.shape[0], step=2):
    pop_corr_trace.append(scipy.stats.spearmanr(population_mean[:i], labels[:i]))

pop_corr_trace = np.array(pop_corr_trace)

# Interclass distance
class_A_activity = run_mean_activity[:, labels == 1.0] # All intervals with label = 1
class_B_activity = run_mean_activity[:, labels == 0.0] # All intervals with label = 0

classvec_A = class_A_activity.mean(axis=-1) # Mean vector for class A
classvec_B = class_B_activity.mean(axis=-1) # Mean vector for class B

sns.set_style("whitegrid")
plt.plot(pop_corr_trace[:,0])
plt.ylabel("r")
plt.xlabel("window")
plt.title("Mean Firing Rate to Class spearman correlation")
plt.show()

print("Final r =", pop_corr_trace[-1][0])
print("Interclass Cosine Distance =", scipy.spatial.distance.cosine(classvec_A,classvec_B))
```

## Weights Analysis

```python
# Changes in weight
trajectories = []
for traj in weights:
    traj = np.array(traj)
#     out_n = eKC_activity.shape[1]
#     print(traj.shape)
#     print(np.prod((traj.shape[0], traj.shape[1]//out_n, out_n)))
#     traj = traj.reshape(traj.shape[0], traj.shape[1]//out_n, out_n)
#     print(traj.shape)
    trajectories.append(traj)
print(len(trajectories))
print(weights.shape)
print(eKC_activity.shape)
```

```python
# Plot statistics
def plot_traj_statistic(f, name, ax, ylabel):
    for i,traj in enumerate(trajectories):
        ax.plot(f(traj), label=f"run-{i+1}")
    
    ax.legend()
    ax.set_title(name)
    ax.set_ylabel(ylabel)
    ax.set_xlabel(f"sampling step (x{WEIGHT_LOG_FREQ}ms)")
```

```python
sns.set_style("whitegrid")
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,6))

fig.suptitle("Weight Statistic Trajectories")
plot_traj_statistic(lambda t: np.mean(t, axis=1), "Mean", ax1, "mean weight (nA)")
plot_traj_statistic(lambda t: np.std(t, axis=1), "Standard Deviation", ax2, r"$\delta$ (nA)")
```

```python
init_dist = []
final_dist = []
for traj in trajectories:
    init_dist.append(traj[0])
    final_dist.append(traj[-1])

init_dist = np.array(init_dist).mean(axis=0)
final_dist = np.array(final_dist).mean(axis=0)

# Plot
fig, ax = plt.subplots(figsize=(10,6))

ax.boxplot([init_dist, final_dist], labels=["initial", "final"], patch_artist=True, sym="")
ax.set_ylabel("weight")
ax.set_title("Weight Distribution")

plt.show()
```
